```python
import scipy
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
import wave
import librosa
import numpy as np
```

```python
samplerate, data = wavfile.read('Welcome.wav')

input_data = wave.open("Welcome.wav", "r")
input_data.getnchannels()
```

```python
print("Data shape: ", data.shape, "\n")
length = data.shape[0] / samplerate
print(length)
time = np.linspace(0., length, data.shape[0])
plt.plot(time, data, label="data")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```

```python
print("samplerate: ", samplerate)
```

```python
frames = []
frame_size = 0.020
overlab_size = 0.005
number_samples_per_frame = int(frame_size * samplerate)
overlab_len = int(overlab_size * samplerate)
step = (number_samples_per_frame - overlab_len)
for i in range(0, len(data), step):
    frame = np.zeros([number_samples_per_frame,])
    frame[0:len(data[i:i+number_samples_per_frame])] = data[i:i+number_samples_per_frame]
    frames.append(frame)
```

```python
print("\nnumber of samples per frame: ", number_samples_per_frame)
print("Len of frames: ", len(frames),"\n")

# for i in range(0, len(frames)):
#   print("fram ", i, ": ", frames[i])
#   print(frames[i].shape)

print("frame 0: ", frames[0], "\n")
print("frame ",len(frames)-1, " : " , frames[len(frames)-1])
```

```python
plt.figure(figsize=(20, 4))
for i in range(10):
  plt.subplot(2, 5, i + 1)
  plt.plot(frames[i], label="frame")
```

```python
signal = []
for frame in frames:
    for item in frame:
        signal.append(item)
```

```python
length = len(signal) / samplerate
time = np.linspace(0., length, len(signal))
plt.plot(time, signal, label="data")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```
